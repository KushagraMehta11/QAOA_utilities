<tt>tsp_generate_coeffs(J_or_G, scale, epsilon, **kwargs)</tt>
<dd>
 generate coefficients for TSP starting at location 0 using (N-1)**2 qubits <br>
     (in matrix form: rows=locations, columns=time steps; ideally initialized with superposition of W-states) <br>
  (based on "Important Quantum Gates for Quantum Algorithms of Travelling Salesman Problem",
    by T.J.H.Sinaga et al, in Proceedings of ICoABCD 2023) <br>
&nbsp; Input arguments (default value in braces for optional arguments):<br>
<tt>J_or_G</tt>:  &nbsp; &nbsp; Ising matrix (distance matrix) or Graph object<br>
<tt>scale</tt>:  &nbsp; &nbsp; scale factor (*|scale|) if scale<0 or normalize to scale value &nbsp; [2*np.pi] <br>
<tt>epsilon</tt>:  &nbsp; &nbsp; discard term if scaled coefficient |coeff|&lt; epsilon &nbsp; [0.0] <br>
<tt>**kwargs</tt>:  &nbsp; &nbsp; other parameters (not used yet)
</dd>
    <br>
    <br>
    (d) <strong>qml_qaoa_utilities</strong>: &nbsp; standard cost and mixer Hamiltonians, QAOA circuit definition, optimizer loop for pennylane framework <br>
<br>

<tt>qml_standard_cost_h(weight_dict, **kwargs) </tt>
<dd>
     returns cost Hamiltonian as sum of terms like coeff<sub>j</sub>*PauliZ(q<sub>a</sub>) and coeff<sub>k</sub>*(PauliZ(q<sub>a</sub>) @ PauliZ(q<sub>b</sub>)) <br>
    &nbsp; Input arguments (default value in braces for optional arguments):<br>
    <tt>weight_dict</tt>: &nbsp; &nbsp; dict of coefficients (weights) of form &nbsp; <tt>{(qa,qb): coeff}</tt>  or  <tt>{(qa): coeff}</tt> <br>
    <tt>**kwargs</tt>: additional parameters (not used yet)
</dd>

<br>
<tt>qml_general_cost_h(obs_weight_dict, **kwargs) </tt>
<dd>
     returns cost Hamiltonian as sum of terms like coef<sub>j</sub>*Obs(q<sub>a</sub>) and coef<sub>k</sub>*(Obs(q<sub>a</sub>) @ Obs(q<sub>b</sub>)) <br>
    &nbsp; Input arguments:<br>
    <tt>obs_weight_dict</tt>: &nbsp; &nbsp; dict of Observables and coefficients (weights) of form &nbsp; <tt>{(Oqa,Oqb): coeff}</tt>  or  <tt>{(Oqa): coeff}</tt> or <tt>{(O_qa,O_qb): coeff}</tt>  or  <tt>{(O_qa): coeff}</tt> &nbsp; (or format like for <qml_standard_cost_h()</tt>, which sets 'O'='Z') <br>
        &nbsp; (here 'qa','qb' are qubit numbers and 'O' operator ('H','X','Y','Z','S','T','V'...) <br>
    <tt>**kwargs</tt>: additional parameters, currently defined: <br>
        &nbsp; skiplist=[...] &nbsp; skip entries with keys like "offset" or "constant" (these are the defaults)<br>
        &nbsp; epsilon=... &nbsp; discard terms with coefficients |coeff|<epsilon
</dd>
 <br>

<tt>qml_standard_mixer_h(nwires) </tt>
<dd>
     returns mixer Hamiltonian as sum of PauliX(qa) for each qubit <br>
    &nbsp; Input argument:<br>
    <tt>nwires</tt>: &nbsp; &nbsp; number of qubits
</dd>
 <br>
<tt>qml_qaoa_initial_h(nwires, topo="x+") </tt>

<tt>qml_qaoa_circuit(params, cost_h, mixer_h, initial_state, nlayers)</tt>
<dd>
define a QAOA circuit of nlayers; &nbsp; <tt>params</tt> &nbsp; and &nbsp; <tt>cost_h</tt> must be provided. <br>
    Note: calling 'qml_qaoa_initial_h' doesn't work yet, so 'qml_qaoa_circuit'
          also initializes the states (see 'initial_state' argument) <br>
    &nbsp; Input arguments (default value in braces for optional arguments):<br>
<tt>params</tt>: &nbsp; &nbsp;  2-dim list of parameters &nbsp;<tt>[gammas, alphas]</tt>&nbsp; for cost and mixer terms for each layer <br>
<tt>cost_h</tt>: &nbsp; &nbsp;  cost Hamiltonian (must be provided)<br>
<tt>mixer_h</tt>: &nbsp; &nbsp;  mixer Hamiltonian (calls 'qml_standard_mixer_h' if None) [None] <br>
<tt>nlayers</tt>: &nbsp; &nbsp;  number of QAOA layers (time steps) &nbsp; [1] <br>
<tt>initial_state</tt>: &nbsp; &nbsp; setting for state initialization  &nbsp; ["x+"] <br>
&nbsp; &nbsp;  <tt>initial_state="x+" or "+"</tt> initialize all qubits to |+&rangle; state  (default) <br>
&nbsp; &nbsp; <tt>initial_state="-" or "x-"</tt>  initialize all qubits to |-&rangle; state <br>
&nbsp; &nbsp; <tt>initial_state="W"</tt>          initialize to superposion of W-states  &nbsp; (W = |100..&rangle; + |010..&rangle; + |001..&rangle; + ...) <br>
&nbsp; &nbsp; <tt>initial_state="rand"</tt>      initialize qubits with random rotation about Y-axis <br>
&nbsp; &nbsp; <tt>initial_state=[..,..,..]</tt>  list with rotation angle for each qubit
</dd>
 <br>

<tt>qml_qaoa_optimize(cost_function, optimizer, initparams_or_nlayers, max_iter, epsilon, printcycle, outdata)</tt>
<dd>
routine to run optimization loop of QAOA circuit ('cost_function'), optimizer class must be provided <br>
&nbsp; Input arguments (default value in braces for optional arguments):<br>
<tt>cost_function</tt>: &nbsp; &nbsp; cost function circuit <br>
<tt>optimizer</tt>: &nbsp; &nbsp; classical optimizer for this task <br>
<tt>initparams_or_nlayers</tt>: &nbsp; &nbsp;  set of initial parameters [gammas,alphas] or number of layers (then all init. parameters are set to 0.5) <br>
<tt>max_iter</tt>: &nbsp; &nbsp;  max. number of iterations (unless |&Delta costs|&lt;epsilon) &nbsp; [100]<br>
<tt>epsilon</tt>: &nbsp; &nbsp;   early termination if |&Delta; costs|&lt;epsilon (epsilon=0 forces all iterations) &nbsp; [0.0]<br>
<tt> printcycle</tt>: &nbsp; &nbsp;  printout every N iterations &nbsp; [10]<br>
<tt> outdata</tt>: &nbsp; &nbsp;     routine returns last parameter set and list of costs for each iteration (if outdata=1) or lists of all parameter sets and costs for each iteration (if outdata=2) (if outdata=-1 or -2 draw additionally the circuit with parameters in this iteration) &nbsp; [1]
</dd>


```python
# imports 
#  (note: it's best to import also the original numpy, under a different name 
#   than pennylane's (autograd's) numpy - some packages require the original
#   numpy classes)
import numpy as np
from matplotlib import pyplot as plt
import networkx as nx
import pennylane as qml
from pennylane import numpy as pnp

from qaoa_utilities import *
from qml_qaoa_utilities import *

# set seed for reproducibility
np.random.seed(seed=1086421)

# create positions (x,y) for N nodes
N=5
pos = np.random.rand(N,2)*100

# create graph for these positions (edges between all nodes will be generated)
graph = create_graph(pos)

# get weight matrix from graph attributes
print("Weight matrix:\n", get_weight_matrix(graph))

# get edges and weights as dict { (nodeA,nodeB): weight} 
#   scale>0 normalize to this value (default: 2*np.pi); scale<0 multiply by |scale|     
print("\nDict of normalized weights:\n",get_normalized_weights(graph, scale=-1))

# plot graph with edge weights (linewidth and values)
plot_graph(graph, plot_weights=True)
```

### Set up the QAOA circuit and run optimization loop

```python
# (a) define the backend device 

# we need the number of qubits (wires) to define the device
nqubits = graph.number_of_nodes()
print("Number of qubits:",nqubits)

# set nshots>0 and qml.device(..., shots=nshots) to get sampled results
nshots = 100

#dev = qml.device('default.qubit', wires=nqubits) #,shots=nshots)
dev = qml.device("qulacs.simulator", wires=nqubits, shots=nshots)
#  for IonQ devices: verbose=1 prints time of submitting and retrieving jobs
#     verbose=2: also print result (json dict); verbose=3: also print submitted json dict
#dev = qml.device("ionq.simulator", target="simulator.aria-1", wires=nqubits, shots=nshots, verbose=0)
#dev = qml.device("ionq.simulator", wires=nqubits, shots=nshots, verbose=1)
# lightning.qubit does not have the standard sampling option (only 'shots=1')!
#dev = qml.device("lightning.qubit", wires=nqubits)
```

```python
# (b) number of QAOA layers 
# (each layer consists of U(gamma_i*cost_h)*U(alpha_i*mixer_h) with i=layer number)
nlayers = 5

# (c) define the cost Hamiltonian
# quantum circuit works best if weights are normalized 
#  (potentially to a smaller range than 0 to 2*pi ... to be studied!)
coeffs = get_normalized_weights(graph)
print(f"Normalized coefficients for cost Hamiltonian ({nlayers} layers):\n",coeffs,"\n")

# cost hamiltonian (must be defined outside of qml_qaoa_circuit)
#    here: use a standard hamiltonian (sum of coef_i*Z(qa) and coef_ij*Z(qa)@Z(qb)
cost_h = qml_standard_cost_h(coeffs)

# (d) define cost function (QAOA circuit)
#   params is list of all gamma_i and alpha_i: [[gamma_0,...],[alpha_0,...]]
#   sample argument to switch between expect.value and probabilities or samples as output
#   (note: pennylane's own backend devices keep the number of shots as dev.shots.total_shots,
#    others as dev.shots)

@qml.qnode(dev)
def cost_function(params, sample=0):
    qml_qaoa_circuit(params, cost_h, initial_state="+", nlayers=nlayers)
    if sample != 0:
        nshots = dev.shots.total_shots if hasattr(dev.shots, 'total_shots') else dev.shots
        if nshots is None: nshots = 0
        return qml.sample(wires=list(range(nqubits))) if sample*nshots>0 else qml.probs(wires=list(range(nqubits)))
    return qml.expval(cost_h)

# (e) set initial parameters for the optimizer 
params = pnp.array(np.repeat(0.5, 2*nlayers).reshape(2,nlayers))

# and optionally draw the cost_function circuit 
#     (expansion_strategy=... shows all gates and rotation angles)
print(qml.draw(cost_function, expansion_strategy="device")(params))

# plot of the cost_function circuit using matplotlib (up to 16 wires)
qml.draw_mpl(cost_function)(params)
```

```python
# (f) choose classical optimizer 
#optimizer = qml.GradientDescentOptimizer()
#optimizer = qml.AdagradOptimizer(stepsize=0.5)
optimizer = qml.AdamOptimizer(stepsize=0.1)
#optimizer = qml.QNGOptimizer(stepsize=0.01, lam=0.001, approx="block-diag")
#optimizer = qml.QNSPSAOptimizer(stepsize=0.01, regularization=0.01)

# (g) run iterations to minimize the cost_function 
#  function:
#    qml_qaoa_optimize(cost_function, optimizer, initparams_or_nlayers, 
#         max_iter=100, epsilon=0., printcycle=10, outdata=1)
#  initparams_or_nlayers: parameters will be set to 0.5 if the number of layers is diven
#  epsilon>0: stop iterations after 3 consecutive iterations with |cost-previous_cost|<epsilon
#  outdata=1: returns best_params, costs_list 
#  outdata=2: returns params_of_all_iterations, costs_list   
# (outdata<0: also draw circuit for every printcyle iteration)
params,costs = qml_qaoa_optimize(cost_function, optimizer, params, \
                                 max_iter=200, epsilon=0.1, printcycle=5)

# (note: running in sampling mode takes very long!)
```

```python
# print costs vs iterations
#  (the simple label "costs" or "Costs" sets also titel and xlabel)
plot_histogram(costs, "costs")
```

```python
# to get probabilities or sampled counts for each state 
# run the QAOA circuit for the final parameters
# set sample=-1 to get probabilities or sample=1 to get an array of 'nshots' samples 
# (if the backend device was not initialized for shots, i.e. nshots=0, then probabilities are returned) 

probs = cost_function(params,sample=1)

# returns a sample of 'nshots' lists (0 or 1 for each qubit)
print("Size of sample array:", probs.shape)
print("The last samples:\n", probs[-10:])

# get the most probable outcomes (states) in form {'state': probability or counts}
#  (to get all states, e.g. transform sample array into counts per state, 
#   set 'nresults' to a large number)
res = get_most_probable_states(probs, 5)
print("Most probable outcomes:",res)

# get sorted probabilities (or counts) for all states
get_most_probable_states(probs,len(probs))
```

```python
# you may plot the most probable states by coloring the corresponding nodes
states = list(res.keys())
for j in range(3): 
    print("Coloring for state",states[j])
    plot_graph(graph,colors=states[j])
```

```python
# frequency plot  
plot_probabilities(probs)

# to show all states on x-axis set: all_states=True
# for all_states=True, you may increase the figure width
plot_probabilities(probs, all_states=True, figsize=(16,10)) 
```

### Another example for maxcut

```python
edges = [ #('A','B',5), ('B','C',7), ('C','D',12),
         ("D","E",14), ("E","F",8), ("F","G",9),
         ("G","H",15), ("H","I",6), ("I","J",9),
         ("H","J",11), ("G","J",5), ("E","J",14) ]
gr=create_graph(edges=edges,seed=1686421)
plot_graph(gr, colors="blue")
```

```python
# set up QAOA circuit 
# (a) define the backend device
nqubits = gr.number_of_nodes()
#dev = qml.device('default.qubit', wires=nqubits)  #,shots=nshots)
#dev = qml.device("qulacs.simulator", wires=nqubits) #, shots=nshots)
dev = qml.device("lightning.qubit", wires=nqubits)

# (b) number of QAOA layers 
nlayers = 6

# (c) define the cost Hamiltonian
coeffs = get_normalized_weights(graph)
cost_h = qml_standard_cost_h(coeffs)

# (d) define cost function (QAOA circuit)
@qml.qnode(dev)
def cost_function(params, sample=0):
    qml_qaoa_circuit(params, cost_h, initial_state="+", nlayers=nlayers)
    if sample != 0:
        nshots = dev.shots.total_shots if hasattr(dev.shots, 'total_shots') else dev.shots
        if nshots is None: nshots = 0
        return qml.sample(wires=list(range(nqubits))) if sample*nshots>0 else qml.probs(wires=list(range(nqubits)))
    return qml.expval(cost_h)

# (e) set initial parameters for the optimizer 
params = pnp.array(np.repeat(0.5, 2*nlayers).reshape(2,nlayers))
# and optionally draw the cost_function circuit 
#print(qml.draw(cost_function, expansion_strategy="device")(params))
qml.draw_mpl(cost_function)(params)
```

```python
# (f) choose classical optimizer 
#optimizer = qml.GradientDescentOptimizer()
#optimizer = qml.AdagradOptimizer(stepsize=0.5)
optimizer = qml.AdamOptimizer(stepsize=0.1)

# (g) run iterations to minimize the cost_function
#  here outdata=2: return costs and array of parameters for all 100 iterations
params,costs = qml_qaoa_optimize(cost_function, optimizer, params, \
                                 max_iter=100, epsilon=0, printcycle=5, outdata=2)
```

```python
print("First and last parameter sets:")
for i in range(3):
    print(f"iter {i}: {params[i]}")
for i in range(len(params)-3, len(params)):
    print(f"iter {i}: {params[i]}")

# params for each iteration is of form [[gammas],[alphas]]
#   reshape and transpose the params so it can be used to overlay multiple plots
parray = np.array(params).reshape(len(params),-1).T
#   and plot gammas for all layers on one plot, same for alphas
glabels = [ 'g'+str(j) for j in range(nlayers) ]
alabels = [ 'a'+str(j) for j in range(nlayers) ]
plot_histogram(parray[:nlayers], label=glabels, title="Cost Hamiltonian parameters for each layer", xlabel="Iteration") 
plot_histogram(parray[nlayers:], label=alabels, title="Mixer Hamiltonian parameters for each layer", xlabel="Iteration") 
```

```python
# Create a graph for given edges

edges = [(0,1), (1,2), (3,0), (2,3), (0,4), (3,4), (2,5), (5,4)]
gr = create_graph(edges=edges)

# plotting requires positions of nodes (nx.spring_layout() is used in the code, so 
#   provide a seed for reproducibility of the plot);
# additional plotting arguments can be set ('width' for edges, the others for nodes)
plot_graph(gr, seed=86421, node_size=2000, font_size=32, font_color="black",  
           node_color="green", width=3)
```

```python
nqubits = gr.number_of_nodes()
print("Number of qubits:",nqubits)
nshots = 1000
#dev = qml.device("lightning.qubit", wires=nqubits)
dev = qml.device("default.qubit", wires=nqubits, shots=nshots)

cost_h = qml_standard_cost_h(get_normalized_weights(gr))
print(cost_h)

myparams = []   #result: final parameter set for each run (nlayers setting)
mycosts = []    #result: final costs for each run
myprobs = []    #result: probabilities per state for each run
mystate = []    #result: most probable state for each run
mylayers= [3, 5, 7, 9]    #list of nlayers to check

for nlayers in mylayers:
    print("\n<<<<<<<<     Number of layers: {}     >>>>>>>>".format(nlayers))

    @qml.qnode(dev)
    def cost_function(params, sample=0):
        qml_qaoa_circuit(params, cost_h, initial_state="+", nlayers=nlayers)
        if sample != 0:
            nshots = dev.shots.total_shots if hasattr(dev.shots, 'total_shots') else dev.shots
            if nshots is None: nshots = 0
            return qml.sample(wires=list(range(nqubits))) if sample*nshots>0 else qml.probs(wires=list(range(nqubits)))
        return qml.expval(cost_h)

    #optimizer = qml.GradientDescentOptimizer()
    #optimizer = qml.AdagradOptimizer(stepsize=0.5)
    optimizer = qml.AdamOptimizer(stepsize=0.1)

    params,costs = qml_qaoa_optimize(cost_function, optimizer, initparams_or_nlayers=nlayers, max_iter=200, epsilon=0.1, printcycle=5)
    myparams.append(params)
    mycosts.append(costs)
    probs = cost_function(params, sample=1)
    devshots = dev.shots.total_shots if hasattr(dev.shots, 'total_shots') else dev.shots
    if devshots is not None and devshots > 0:
        print("sampling mode")
        probs = get_most_probable_states(probs, len(probs))
    myprobs.append(probs)
    best_states = get_most_probable_states(probs, nresults=4)
    mystate.append(list(best_states.keys())[0])
    print("Most probable states:", best_states)
    del cost_function
    del optimizer
            
```

```python
# show probabilities and best solutions for each 'nlayers' setting
for i,nlayers in enumerate(mylayers): 
    title = f"nlayers={nlayers}" 
    plot_probabilities(myprobs[i], title=title, all_states=True)
    plot_graph(gr,colors=mystate[i], seed=86421)
```
